```python
import pandas as pd

df = pd.read_excel('../data/online_retail.xlsx')

print(df.head())
```

```python

# Rows and columns
print("\nShape:", df.shape)

# Data types
print("\nData Types:")
print(df.dtypes)

# Missing values
print("\nMissing Values:")
print(df.isnull().sum())

for col in df.columns:
    print(f"{col}: {df[col].duplicated().sum()}")

negative_counts = (df.select_dtypes(include='number') < 0).sum()

print(negative_counts)
```

```python
total_rows = len(df)
duplicate_rows = df.duplicated().sum()

print(f"Total rows: {total_rows}")
print(f"Duplicate rows: {duplicate_rows}")
print(f"Percentage: {duplicate_rows/total_rows*100:.2f}%")
```

```python
print("Rows before cleaning:", len(df))
print("Columns before cleaning:", len(df.columns))
```

```python
# Remove missing descriptions
df = df.dropna(subset=['Description'])

# Remove missing customer IDs
df = df.dropna(subset=['Customer ID'])

# Remove returns
df = df[df['Quantity'] > 0]

# Remove invalid prices
df = df[df['Price'] > 0]

# Remove exact duplicate rows
df = df.drop_duplicates()
```

```python
print("Rows after cleaning:", len(df))
print("Columns after cleaning:", len(df.columns))
```

```python
df['Revenue'] = df['Quantity'] * df['Price']
```

```python
print(df[['Quantity', 'Price', 'Revenue']].head())
```

```python
# Extract date components
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month_name()
df['DayOfWeek'] = df['InvoiceDate'].dt.day_name()

# Check results
print(df[['InvoiceDate', 'Year', 'Month', 'DayOfWeek']].head())
```

```python
df.to_csv('../data/online_retail_cleaned.csv', index=False)

print("File saved successfully!")
```
